```python
def time_to_value(x):
    time, period = x.split(' ')
    hours, minutes = time.split(':')
    hours, minutes = int(hours), int(minutes)

    if hours == 12:    hours = 0
    if period == 'PM': hours += 12
    minutes /= 100

    return hours + minutes

childcare_df['start_hours_val'] = childcare_df['hours_of_operation'].apply(lambda x: time_to_value(x.split('-')[0]))
childcare_df['end_hours_val'] =   childcare_df['hours_of_operation'].apply(lambda x: time_to_value(x.split('-')[1]))
```

```python
def age_array(x):
    x = x.split(',')
    x = ["'" + ele + "'" for ele in x]
    x = ','.join(x)
    x = '[' + x + ']'
    x = eval(x)
    x = list(x)
    return x

childcare_df['licensed_to_serve_ages'] = childcare_df['licensed_to_serve_ages'].apply(lambda x: age_array(x))
childcare_df['days_of_operation'] = childcare_df['days_of_operation'].apply(lambda x: age_array(x))
```

```python
childcare_df['start_hours_val'] = childcare_df['start_hours_val'].apply(lambda x : int(round(x)))
childcare_df['end_hours_val'] = childcare_df['end_hours_val'].apply(lambda x : int(round(x)))
```

```python
def get_zipcode(x):
    if x[-5] == ' ':
        zipcode = x.split(' ')[-2]
    else:
        zipcode = x.split(' ')[-1]

    return int(zipcode)

childcare_df['zip_code'] = childcare_df['location_address'].apply(lambda x: get_zipcode(x))
```
